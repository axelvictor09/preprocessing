```python
df=pd.read_csv("/content/preprocessing_practice_dataset.csv")
```

```python
df['Age_mean_imputed'] = df['Age'].fillna(df['Age'].mean())
df['Age_median_imputed'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Age_mean_imputed'] = df['Age'].fillna(df['Age'].mean())
df['monthly_mean_imputed'] = df['Monthly_Salary'].fillna(df['Monthly_Salary'].mean())
df['education_mode_imputed'] = df['Education'].fillna(df['Education'].mode()[0])
df['city_mode_imputed'] = df['City'].fillna(df['City'].mode()[0])
df['dep_mode_imputed'] = df['Department'].fillna(df['Department'].mode()[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Employee_ID',y='Department',palette='pastel')
plt.title('Boxplot of Department by Employee id')
plt.xlabel('Employee id')
plt.ylabel('Department')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x='Employee_ID',y='Department',palette='pastel')
plt.title('Boxplot of Department by Employee id')
plt.xlabel('Employee id')
plt.ylabel('Department')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Employee_ID',y='Department',palette='pastel')
plt.title('Boxplot of Department by Employee id')
plt.xlabel('Employee id')
plt.ylabel('Department')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Employee_ID',y='Department',palette='pastel')
plt.title('Boxplot of Department by Employee id')
plt.xlabel('Employee id')
plt.ylabel('Department')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
department_counts = df['Department'].value_counts()
colors = sns.color_palette('pastel')[0:len(department_counts)]
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Distribution of Departments')
plt.ylabel('')
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
file_path='/content/preprocessing_practice_dataset.csv'
p_data=pd.read_csv(file_path)
numerical_columns=['Age', 'Experience_Years', 'Monthly_Salary', 'Performance_Rating']
p_data=p_data.dropna(subset=numerical_columns)
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
p_data['p_minmax']=minmax_scaler.fit_transform(p_data[['Performance_Rating']])
p_data['e_standard']=minmax_scaler.fit_transform(p_data[['Experience_Years']])
p_data['m_standard']=standard_scaler.fit_transform(p_data[['Monthly_Salary']])
p_data['a_standard']=standard_scaler.fit_transform(p_data[['Age']])
print(p_data[['Age','a_standard','Performance_Rating','p_minmax','Experience_Years','e_standard','Monthly_Salary','m_standard']])
```
